The following dataset that we will be examining are measurements in order to identify Higgs bosons at high transverse momenta, which decay to bottom-quark pairs. These measurements were made at the Large Hadron Collider.

This is an extension of the ATLAS experiment, of proton-proton collisions, to heavy new particles which may have decay chains that include Higgs bosons. The large mass-splitting between these heavier particles and their decay products results in high-momenta Higgs boson. Due to high-momenta, the decay products come off collimated (or in other words, coming off parallel to each other).

The particle collisions occur at a center-of-mass energy of 13 TeV. The ATLAS detector covers nearly the entire solid angle around the collision, with an inner tracking detector surrounded by a thin superconducting solenoid electromagnetic and hadronic calorimeter, and a muon spectrometer incorporating three large superconducting toroid magnets.

There are 14 different measurements in our dataset. These are $\rho$<sub>t</sub>, $\eta$, $\phi$, m, EE2, EE3, d<sub>2</sub>, angularity, $\tau$<sub>1</sub>, $\tau$<sub>2</sub>, $\tau$<sub>3</sub>, $\tau$<sub>21</sub>, $\tau$<sub>32</sub>, K<sub>t</sub>Î”R:

$\rho$<sub>t</sub> represents the transverse momentum of the particle.<br>
$\eta$ represents the pseudorapidity. The angle between the axis which runs through the path of particles colliding, and the vertical axis.<br>
$\phi$ represents the azimuthal angle, which is the angle between the axis which runs across the ATLAS detector and the vertical axis.<br>
m is the invariant mass. The portiion of the total mass that is independent of the total motion of the particle.<br>

EE2 and EE3 are the 2 and 3-point ECF ratios. The ratios generally represent how far the particles are within a jet.

angularity

$\tau$<sub>1</sub>, $\tau$<sub>2</sub>, $\tau$<sub>3</sub>, $\tau$<sub>21</sub> are the values which represent how likely the measured particles are a 1, 2, or 3 prong jet. The $\tau$<sub>21</sub>, $\tau$<sub>32</sub> represent the ratios between these values; $\tau$<sub>21</sub> is the ratio between $\tau$<sub>2</sub> and $\tau$<sub>2</sub>

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import scipy
from scipy import stats
import pickle

infile = open ("higgs_100000_pt_250_500.pkl",'rb')
higgs = pickle.load(infile)

higgs.keys()
```

```python
infile = open ("qcd_100000_pt_250_500.pkl", 'rb')
qcd = pickle.load(infile)

qcd.keys()
```

Below the invariant mass and transverse momentum from the Higgs and QCD dataset is plotted onto histograms. We can see that for both plots, the Higgs dataset has an extremely tall and narrow peak, while the QCD data set is flatter. The peak from mass of Higgs dataset can be explained due to Heisenburg's Uncertainty Principle, which also explains the peak for the momentum plot.

```python
plt.rcParams["figure.figsize"] = (15,6)
grid = plt.GridSpec(1, 2, wspace=1, hspace=0.5)

plt.subplot(grid[0, :1])
pt_higgs = higgs['pt']
pt_plot = plt.hist(pt_higgs, bins=250, histtype='step', label='Higgs')
pt_qcd = qcd['pt']
plot = plt.hist(pt_qcd, bins=250, histtype='step', label='QCD')
plt.title('Higgs/QCD Momentum Data')
plt.xlabel('Momentum')
plt.ylabel('Count')
plt.legend(prop={'size': 10})

plt.subplot(grid[0, 1:])
pt_higgs = higgs['mass']
pt_plot = plt.hist(pt_higgs, bins=250, histtype='step', label='Higgs')
pt_qcd = qcd['mass']
plot = plt.hist(pt_qcd, bins=250, histtype='step', label='QCD')
plt.title('Higgs/QCD Mass Data')
plt.xlabel('Momentum')
plt.ylabel('Count')

plt.legend(prop={'size': 10})
```

```python
#0 means the particle shoots off in the y direction (upwards). We're more interested in the range of roughly -2
#to 2. 3 represents the particle shoots off in the z direction, which means the colliding particles grazed each
#other.

pt_higgs = higgs['eta']
pt_plot = plt.hist(pt_higgs, bins=250, histtype='step')

pt_qcd = qcd['eta']
plot = plt.hist(pt_qcd, bins=250, histtype='step')
```
